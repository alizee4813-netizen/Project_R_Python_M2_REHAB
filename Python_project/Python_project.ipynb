```python
# 1. Import libraries
import pandas as pd
import os
```

# 1. Import the Datasets

In this first step, the two datasets are imported into Python.
The `GAITERITE.csv` file contains gait data collected from the GAITERITE system.
The `RANDO.csv` file contains the randomization information, meaning the group assignment for each patient.
This step is necessary because the information required for the analysis is distributed across two different files.

```python
# Import walking data
gait = pd.read_csv("data/GAITERITE.csv", sep=";", encoding="latin1")

# Import randomisation data
rando = pd.read_csv("data/RANDO.csv", sep=";", encoding="latin1")

# Displaying the First Rows to Verify the Import
display(gait.head())
display(rando.head())
```

# 2. Checking the Columns

Before merging the two datasets, it is important to verify the column names present in each file.
This step helps identify the common column that will be used to link the gait data with the randomization data.

```python
print("Columns GAITERITE :")
print(gait.columns.tolist())

print("\nColumns RANDO :")
print(rando.columns.tolist())
```

# 3. Standardizing Patient Identifier Formats

Both datasets contain a patient identifier, but the column names are not exactly the same: `NumPAT` in the `GAITERITE` dataset and `Numpat` in the `RANDO` dataset.
In addition, Python cannot merge two columns if they are not in the same format.
Therefore, both identifiers are converted into text format to enable the merge.

```python
# Converting patient identifiers into text format
gait["NumPAT"] = gait["NumPAT"].astype(str)
rando["Numpat"] = rando["Numpat"].astype(str)
```

# 4. Merging the Two Datasets

The `GAITERITE` and `RANDO` datasets are merged using the patient identifier.
This step makes it possible to obtain a single dataset containing both:
the gait parameters,
and the randomization group of each patient.
The merge is essential in order to compare the mean double support time between the groups.

```python
# Merging the two datasets
df = pd.merge(gait, rando, left_on="NumPAT", right_on="Numpat")

# Checking the merge
df.head()
```

# 5. Selecting Valid Data

Not all rows in the dataset are necessarily usable.
The `COMMENT` column indicates whether a measurement is valid or not.
In this step, only observations labeled `VALIDE` are retained.
This selection ensures that the analysis is performed only on reliable data and prevents incorrect measurements from influencing the results.

```python
# Select valid observations
df = df[df["COMMENT"] == "VALIDE"]

# Checking after filtering
df.head()
```

# 6. Creating the main variable: Mean Double Support Time

Double support time is measured separately for the left and right sides.
To obtain an overall measure of stability, both values are combined into a single average variable called `DoubleSupport_mean`.
Before the calculation, the columns are converted into numeric format because some values may be read as text due to the use of commas as decimal separators.

```python
# Converting the left double support column to numeric
df["DoubleSupport_G"] = pd.to_numeric(
    df["Durée du double appui (s)GAUCHE"].astype(str).str.replace(",", "."),
    errors="coerce"
)

# Converting the right double support column to numeric
df["DoubleSupport_D"] = pd.to_numeric(
    df["Durée du double appui (s)DROIT"].astype(str).str.replace(",", "."),
    errors="coerce"
)

# Calculating the mean double support time
df["DoubleSupport_mean"] = (
    df["DoubleSupport_G"] + df["DoubleSupport_D"]
) / 2

# Checking the new variable
df[["NumPAT", "RANDOMISATION", "DoubleSupport_G", "DoubleSupport_D", "DoubleSupport_mean"]].head()
```

# 7. Exporting the Clean Dataset

The final dataset is exported under the name `results.csv`.
This file contains the cleaned data, the randomization variable, and the main variable DoubleSupport_mean.
It will then be used in R to perform the descriptive analyses and calculate the effect size.

```python
# Exporting the clean dataset
df.to_csv("data/results.csv", index=False)
```
