# 1. Import des données

Dans cette première étape, les deux bases de données sont importées dans Python.

Le fichier `GAITERITE.csv` contient les données de marche issues du système GAITERITE.  
Le fichier `RANDO.csv` contient les informations de randomisation, c’est-à-dire le groupe auquel appartient chaque patient.

Cette étape est nécessaire car les informations utiles à l’analyse sont réparties dans deux fichiers différents.

```python
# 1. Import des bibliothèques
import pandas as pd
import os
```

```python
# Importation des données de marche
gait = pd.read_csv("data/GAITERITE.csv", sep=";", encoding="latin1")

# Importation des données de randomisation
rando = pd.read_csv("data/RANDO.csv", sep=";", encoding="latin1")

# Affichage des premières lignes pour vérifier l'importation
display(gait.head())
display(rando.head())
```

# 2. Vérification des colonnes

Avant de fusionner les deux bases, il faut vérifier le nom des colonnes présentes dans chaque fichier.

Cette étape permet d’identifier la colonne commune qui servira à associer les données de marche avec les données de randomisation.

```python
print("Colonnes GAITERITE :")
print(gait.columns.tolist())

print("\nColonnes RANDO :")
print(rando.columns.tolist())
```

# 3. Mise au même format des identifiants patients

Les deux fichiers possèdent un identifiant patient, mais les noms des colonnes ne sont pas exactement les mêmes : `NumPAT` dans la base GAITERITE et `Numpat` dans la base RANDO.

De plus, Python ne peut pas fusionner deux colonnes si elles ne sont pas au même format.  
Les deux identifiants sont donc transformés en texte afin de permettre la fusion.

```python
# Transformation des identifiants patients en texte
gait["NumPAT"] = gait["NumPAT"].astype(str)
rando["Numpat"] = rando["Numpat"].astype(str)
```

# 4. Fusion des deux bases de données

Les données GAITERITE et RANDO sont fusionnées à partir de l’identifiant patient.

Cette étape permet d’obtenir une seule base contenant à la fois :
- les paramètres de marche ;
- le groupe de randomisation de chaque patient.

La fusion est indispensable pour pouvoir comparer le temps de double appui moyen entre les groupes.

```python
# Fusion des deux bases
df = pd.merge(gait, rando, left_on="NumPAT", right_on="Numpat")

# Vérification de la fusion
df.head()
```

# 5. Sélection des données valides

Toutes les lignes du fichier ne sont pas forcément exploitables.  
La colonne `COMMENT` indique si une mesure est valide ou non.

Dans cette étape, seules les observations marquées `VALIDE` sont conservées.

Cette sélection permet de travailler uniquement sur des données fiables et d’éviter que des mesures incorrectes influencent les résultats.

```python
# Sélection des observations valides
df = df[df["COMMENT"] == "VALIDE"]

# Vérification après filtrage
df.head()
```

# 6. Création de la variable principale : temps de double appui moyen

Le temps de double appui est mesuré séparément à gauche et à droite.

Pour obtenir une mesure globale de stabilité, les deux valeurs sont combinées en une seule variable moyenne : `DoubleSupport_mean`.

Avant le calcul, les colonnes sont converties en format numérique, car certaines valeurs peuvent être lues comme du texte à cause des virgules utilisées comme séparateurs décimaux.

```python
# Conversion de la colonne double appui gauche en numérique
df["DoubleSupport_G"] = pd.to_numeric(
    df["Durée du double appui (s)GAUCHE"].astype(str).str.replace(",", "."),
    errors="coerce"
)

# Conversion de la colonne double appui droit en numérique
df["DoubleSupport_D"] = pd.to_numeric(
    df["Durée du double appui (s)DROIT"].astype(str).str.replace(",", "."),
    errors="coerce"
)

# Calcul du double appui moyen
df["DoubleSupport_mean"] = (
    df["DoubleSupport_G"] + df["DoubleSupport_D"]
) / 2

# Vérification de la nouvelle variable
df[["NumPAT", "RANDOMISATION", "DoubleSupport_G", "DoubleSupport_D", "DoubleSupport_mean"]].head()
```

# 7. Export de la base propre

La base finale est exportée sous le nom `results.csv`.

Ce fichier contient les données nettoyées, la variable de randomisation et la variable principale `DoubleSupport_mean`.

Il sera ensuite utilisé dans R pour réaliser les analyses descriptives et le calcul de la taille d’effet.

```python
# Export de la base propre
df.to_csv("data/results.csv", index=False)
```
